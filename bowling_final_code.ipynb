```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("mostwickets.csv")
b = pd.read_csv("best_bowing_avgs.csv")
c = pd.read_csv("bowlingSR.csv")
d = pd.read_csv("bowlingeconomy.csv")
e = pd.read_csv("most_maidens.csv")
f = pd.read_csv("mostdots.csv")
g = pd.read_csv("mostrunconceded.csv")
h = pd.read_csv("mostvaluableplayer.csv")
```

```python
a.head()
```

```python
data = [a['Player'],e[' Maid'],f[' dots'],h[' Catches'],h[' Run outs'],h[' Stumpings'],h[' Pts']]
headers = ["Player","Maid", "Dots","Catches","Run Outs","Stumping","Pts"]

df_merged = pd.concat(data, axis=1, keys=headers)
```

```python
data_1 = [e['Player'],e[' Maid']]
headers = ["Player","Maid"]
df1_merged = pd.concat(data_1, axis=1, keys=headers)
df1_merged.head()
```

```python
data_2 = [f['Player'],f[' dots']]
headers = ["Player","Dots"]
df2_merged = pd.concat(data_2, axis=1, keys=headers)
df2_merged.head()
```

```python
data_3 = [h['Player'],h[' Catches'],h[' Run outs'],h[' Stumpings'],h[' Pts']]
headers = ["Player","Catches","Run Outs","Stumping","Pts"]
df3_merged = pd.concat(data_3, axis=1, keys=headers)
df3_merged.head()
```

```python
df = pd.merge(df1_merged, df2_merged, on='Player', how='outer')
```

```python
df_f = pd.merge(df, df3_merged, on='Player', how='outer')
```

```python
df_final = pd.merge(a, df_f, on='Player', how='outer')
```
